Grid Search
----

Función que devuelve un esqueleto de modelo

```
def get_model(n_units = 128, dropout = 0.1, lr= 3e-5):
  regressor = Sequential()
  regressor.add(LSTM(units = n_units, return_sequences = True, input_shape = (X_train.shape[1], 1) ))
  regressor.add(Dropout(dropout))
  regressor.add(LSTM(units = n_units, return_sequences = True ))
  regressor.add(Dropout(dropout))
  regressor.add(LSTM(units = n_units))
  regressor.add(Dropout(dropout))
  regressor.add(Dense(units = 1))
  regressor.compile(optimizer = optimizers.Adam(lr=lr), loss = 'binary_crossentropy', metrics= ['binary_crossentropy', 'accuracy'])
  return regressor
```

Parámetros para el Grid Search

```
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.pipeline import Pipeline

kears_estimator = KerasClassifier(build_fn=get_model, verbose=1)

estimator = Pipeline([("kc", KerasClassifier(build_fn=get_model, verbose=1))])

# define the grid search parameters
param_grid = {
    'kc__epochs': [40, 60],
    'kc__n_units': [128, 256],
    'kc__batch_size':[64, 128],
    'kc__lr':[3e-5, 3e-6],
    # 'kc__optimizer':[optimizers.RMSprop, optimizers.Adam, optimizers.sgd],
    'kc__dropout': [0.4, 0.2]
}
```

Preparo los datos en una separación inicial de train/test y añado una columna para que el Grid sepa qué rows van para entrenar y cuales para validar

```
from sklearn.model_selection import PredefinedSplit

# Usando una columna para indicar el conjunto al que pertenece el dato: -1 train, 0 para test/val
X_aux = data_scaled.copy()
X_aux = X_aux.reset_index()
X_aux = X_aux.drop(columns=['Date'])
X_aux = X_aux.reset_index()
# X_aux = X_aux.drop(columns=['Date'])
X_aux['Train_Test'] = np.zeros(len(df))

# indico los límites de train/test 
train_size = int(len(data_scaled) * 0.95)
test_size = len(data_scaled) - train_size
X_aux.loc[0:train_size, ['Train_Test']] = -1
X_aux.loc[train_size:len(data_scaled), ['Train_Test']] = 0

# esta clase permite almacenar las filas que van a ir a train (-1) y las que van a validacion/test (0)
ps = PredefinedSplit(test_fold=X_aux[['Train_Test']])
print(X_aux[X_aux.Train_Test == -1].shape)
print(X_aux[X_aux.Train_Test == 0].shape)
X_aux = X_aux.drop(columns=['Train_Test'])
print(X_aux.head())

# Separo columnas de train y target
X = X_aux.loc[:, X_aux.columns != 'Target'].to_numpy()
y = X_aux.loc[:, X_aux.columns == 'Target'].to_numpy()
print(X.shape, y.shape)

# Y reshape para que el modelo lo lea bien
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
print(X.shape, y.shape)

# train y test por separado
x_training = X_aux.loc[0:train_size, X_aux.columns != 'Target'].to_numpy()
y_training = X_aux.loc[0:train_size, X_aux.columns == 'Target'].to_numpy()

x_testing = X_aux.loc[train_size:len(data_scaled), X_aux.columns != 'Target'].to_numpy()
y_testing = X_aux.loc[train_size:len(data_scaled), X_aux.columns == 'Target'].to_numpy()

x_training = np.reshape(x_training, (x_training.shape[0], x_training.shape[1], 1))
x_testing = np.reshape(x_testing, (x_testing.shape[0], x_testing.shape[1], 1))
```

El Grid Search! A más parámetros, más combinaciones y más tiempo le llevará ejecutar todo

```
grid = GridSearchCV(estimator=estimator,  
                    n_jobs=-1, 
                    verbose=1,
                    cv=ps,
                    return_train_score=True,
                    param_grid=param_grid,)

grid_result = grid.fit(X, y) 

# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
